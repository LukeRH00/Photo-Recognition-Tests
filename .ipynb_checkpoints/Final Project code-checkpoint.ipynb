# <br><br>Create training and validation functions to use on models

```python
# For training model at start of epoch
def trainingFunction(cnn_model, d_loader, l_function, opt, ep, write):
    cnn_model.train()
    tr_total_loss = 0
    tr_correct = 0
    
    #loop through all batches
    for batch, (data, label) in enumerate(d_loader):
        data = data.to(device)
        label = label.to(device)
        opt.zero_grad()
        
        # Generate output, get error
        estimate = cnn_model(data)
        tr_loss = l_function(estimate, label)
        
        tr_total_loss += tr_loss.item()
        tr_correct += (estimate.argmax(1) == label).type(torch.float).sum().item()
        
        # Backpropagate and update
        tr_loss.backward()
        optimizer.step()
    
    #calculate percentage of correct for accuracy
    tr_correct /= len(d_loader.dataset)
    tr_avg_loss = tr_total_loss / len(d_loader)
    
    #write intermediate results to tensorboard
    write.add_scalar("Loss/training", tr_avg_loss, ep)
    write.add_scalar("Accuracy/training", 100*tr_correct, ep) #record percentage correct during training
```

```python
# For running validation set at end of epoch 
def validationFunction(cnn_model, d_loader, l_function, ep, write):
    cnn_model.eval()
    v_loss = 0
    v_correct = 0
    with torch.no_grad():
        #loop through all batches
        for data, label in d_loader:
            data = data.to(device)
            label = label.to(device)
            
            #get predicted output, add loss and number correct to totals
            estimate = cnn_model(data)
            v_loss += l_function(estimate, label).item()
            v_correct += (estimate.argmax(1) == label).type(torch.float).sum().item()
            
    #calculate percentage of correct for accuracy
    v_correct /= len(d_loader.dataset)
    v_avg_loss = v_loss / len(d_loader)
    
    
    
    #write intermediate results to tensorboard
    
    #For both of these I forgot to update the writes before I had already trained multiple models, and
    #rather than retrain all of the models to get correctly formatted intermediate results, I left it as is.
    #Incorrectly writes %correct as a decimal (instead of 100*v_correct) and loss as total (instead of v_avg_loss)
    write.add_scalar("Accuracy/validation", v_correct, ep)
    write.add_scalar("Loss/validation", v_loss, ep)
    #print out validation loss for epoch
    print("Total Validation Loss:", v_loss)
```

```python
#used to run training and validation functions for given number of epochs
def trainingLoop(cnn_model, tr_load, val_load, l_fn, opt, epoch, write):
    for t in range(epoch):
        print("\nEpoch:", t+1, "\n-------------------------------")
        trainingFunction(cnn_model, tr_load, l_fn, opt, t+1, write)
        validationFunction(cnn_model, val_load, l_fn, t+1, write)
```

# <br><br><br>Define a basic MLP model

```python
# Define MLP model
# 3 FC layers with the last being equivalent to SoftMax(10)
class basicMLP(nn.Module):
    def __init__(self, num_channel, input_h, input_w, out_num):
        super(basicMLP, self).__init__()
        self.layers = nn.Sequential(
            nn.Flatten(),
            nn.Linear(num_channel*input_h*input_w, 576),
            nn.ReLU(),
            nn.Linear(576, 484),
            nn.ReLU(),
            nn.Linear(484, out_num)
        )

    def forward(self, d):
        out = self.layers(d)
        return out
```

### Set loss function for training

```python
#Here is the cell for setting learning rates and loss functions if needed

loss_function = nn.CrossEntropyLoss()
#loss_function = nn.KLDivLoss()


#set different learning rates
#l_rate = 1e-3
#l_rate = 5e-3
#l_rate = 1e-2
#l_rate = 5e-2
l_rate = 1e-1
```

# <br>  <br><br><br><br><br><br><br> Design a simple CNN model

```python
# Define a simple CNN model

#2 Convolution + MaxPool layers followed by a FC with 10 outputs
class simpleCNN(nn.Module):
    def __init__(self, input_channels, input_h, out_num):
        super(simpleCNN, self).__init__()
        self.layers = nn.Sequential(
            nn.Conv2d(input_channels, 4, (3, 3), (1, 1), (1, 1)),
            nn.BatchNorm2d(4, 1e-05, 0.1, True, True),
            nn.ReLU(True),
            nn.MaxPool2d(2, 2, 0, 1, ceil_mode=False),
            nn.Conv2d(4, 4, (3, 3), (1, 1), (1, 1)),
            nn.BatchNorm2d(4, 1e-05, 0.1, True, True),
            nn.ReLU(True),
            nn.MaxPool2d(2, 2, 0, 1, ceil_mode=False),
            nn.Flatten(),
            nn.Linear(int((input_h**2)/4), out_num)
        )

    def forward(self, d):
        out = self.layers(d)
        return out
```

# <br> <br><br><br><br><br><br><br><br><br><br><br><br><br><br><br> define architecture for a LeNet model

## Version 1 does not have batch normalization after each convolution

```python
# Define a LeNet model

#2 conv + MaxPool layers, 3 FC outputting 10
class LeNet(nn.Module):
    def __init__(self, input_channels, input_h, out_num):
        super(LeNet, self).__init__()
        self.layers = nn.Sequential(
            nn.Conv2d(input_channels, 6, (5, 5), 1, 0),
            #nn.BatchNorm2d(6),
            nn.ReLU(),
            nn.MaxPool2d(2, 2, 0),
            nn.Conv2d(6, 16, (5, 5), 1, 0),
            #nn.BatchNorm2d(16),
            nn.ReLU(),
            nn.MaxPool2d(2, 2, 0),
            nn.Flatten(),
            nn.Linear(int(int((((input_h-4)/2)-4)/2)**2)*16, 120),
            nn.ReLU(),
            nn.Linear(120, 84),
            nn.ReLU(),
            nn.Linear(84, out_num)
        )

    def forward(self, d):
        out = self.layers(d)
        return out
```

## Version 2 (marked as r in saved models) has batch normalization after each convolution

```python
# Define a LeNet model

#2 conv + MaxPool layers, 3 FC outputting 10
class LeNet(nn.Module):
    def __init__(self, input_channels, input_h, out_num):
        super(LeNet, self).__init__()
        self.layers = nn.Sequential(
            nn.Conv2d(input_channels, 6, (5, 5), 1, 0),
            nn.BatchNorm2d(6),
            nn.ReLU(),
            nn.MaxPool2d(2, 2, 0),
            nn.Conv2d(6, 16, (5, 5), 1, 0),
            nn.BatchNorm2d(16),
            nn.ReLU(),
            nn.MaxPool2d(2, 2, 0),
            nn.Flatten(),
            nn.Linear(int(int((((input_h-4)/2)-4)/2)**2)*16, 120),
            nn.ReLU(),
            nn.Linear(120, 84),
            nn.ReLU(),
            nn.Linear(84, out_num)
        )

    def forward(self, d):
        out = self.layers(d)
        return out
```

# <br><br><br><br><br><br><br><br><br><br><br><br><br><br><br><br> define architecture for a VGG-16 model

```python
# Define a VGG-16 model 

#2 Conv x 2 + MaxPool, 2 Conv x 3 + MaxPool, 1 Conv x 3, 4 FC outputting 10
class VGG16(nn.Module):
    def __init__(self, input_channels, input_h, out_num):
        super(VGG16, self).__init__()
        
        if(input_h == 32):
            self.num = 2048
        else:
            self.num = 512
        
        self.layers = nn.Sequential(
            nn.Conv2d(input_channels, 64, (3, 3), 1, 'same'),
            nn.BatchNorm2d(64),
            nn.ReLU(),
            nn.Conv2d(64, 64, (3, 3), 1, 'same'),
            nn.BatchNorm2d(64),
            nn.ReLU(),
            nn.MaxPool2d(2, 2, 0), #for 28x28, new size is 14. for 32x32, new is 16
            nn.Conv2d(64, 128, (3, 3), 1, 'same'),
            nn.BatchNorm2d(128),
            nn.ReLU(),
            nn.Conv2d(128, 128, (3, 3), 1, 'same'),
            nn.BatchNorm2d(128),
            nn.ReLU(),
            nn.MaxPool2d(2, 2, 0), #for 14x14 --> 7. for 16x16 --> 8
            nn.Conv2d(128, 256, (3, 3), 1, 'same'),
            nn.BatchNorm2d(256),
            nn.ReLU(),
            nn.Conv2d(256, 256, (3, 3), 1, 'same'),
            nn.BatchNorm2d(256),
            nn.ReLU(),
            nn.Conv2d(256, 256, (3, 3), 1, 'same'),
            nn.BatchNorm2d(256),
            nn.ReLU(),
            nn.MaxPool2d(2, 2, 0), #7x7 --> 3. 8x8 --> 4
            nn.Conv2d(256, 512, (3, 3), 1, 'same'),
            nn.BatchNorm2d(512),
            nn.ReLU(),
            nn.Conv2d(512, 512, (3, 3), 1, 'same'),
            nn.BatchNorm2d(512),
            nn.ReLU(),
            nn.Conv2d(512, 512, (3, 3), 1, 'same'),
            nn.BatchNorm2d(512),
            nn.ReLU(),
            nn.MaxPool2d(2, 2, 0), #3x3 --> 1. 8x8 --> 2
            nn.Conv2d(512, 512, (3, 3), 1, 'same'),
            nn.BatchNorm2d(512),
            nn.ReLU(),
            nn.Conv2d(512, 512, (3, 3), 1, 'same'),
            nn.BatchNorm2d(512),
            nn.ReLU(),
            nn.Conv2d(512, 512, (3, 3), 1, 'same'),
            nn.BatchNorm2d(512),
            nn.ReLU(),
            nn.Flatten(), #either 512 or 2048
            nn.Linear(self.num, self.num),
            nn.ReLU(),
            nn.Linear(self.num, self.num),
            nn.ReLU(),
            nn.Linear(self.num, self.num),
            nn.ReLU(),
            nn.Linear(self.num, out_num)
        )

    def forward(self, d):
        out = self.layers(d)
        return out
```

# <br><br><br><br><br><br><br><br><br><br><br><br><br><br><br><br> define architecture for a ResNet-18 model

### Needed to define structure of an individual block to be used in resNet. I tried writing the entire resNet architecture within the model definition alone and it kept returning errors

```python
#block definition below allows for 2 types: either the input is not resized so down=None is used and the inputs can combine 
#without problem, OR input is shrunken throughout the sequential layers, so an input for down is needed. Happens when sType > 1

class resNetBlock(nn.Module):
    def __init__(self, input_channels, sType, out_num, down=None):
        super(resNetBlock, self).__init__()
        self.startLayers = nn.Sequential(
            nn.Conv2d(input_channels, out_num, (3, 3), sType, 1),
            nn.BatchNorm2d(out_num),
            nn.ReLU(),
            nn.Conv2d(out_num, out_num, (3, 3), 1, 1),
            nn.BatchNorm2d(out_num)
        )
        self.down = down
        self.lastLayer = nn.Sequential(
            nn.ReLU()
        )
    
    def forward(self, d):
        d2 = d
        d = self.startLayers(d)
        if self.down != None:
            d2 = self.down(d2)
        d = d + d2
        out = self.lastLayer(d)
        return out
```

```python
# Define a ResNet-18 model
# I analyzed the structure of the ResNet in the torchvision library and attempted to follow a similar logic
# Slight differences probably exist, but I tried to make this version as close as possible with my current understanding

class resNet(nn.Module):
    def __init__(self, input_channels, out_num):
        super(resNet, self).__init__()
        
        #first blocks that don't change size
        self.frontBlock = resNetBlock(64, 1, 64)
        
        #2nd set the shrink data
        self.block_21 = resNetBlock(64, 2, 128, nn.Sequential(
            nn.Conv2d(64, 128, (1, 1), 2, 0),
            nn.BatchNorm2d(128)
        ))
        self.block_22 = resNetBlock(128, 1, 128)
        
        #3rd set that shrinks data
        self.block_31 = resNetBlock(128, 2, 256, nn.Sequential(
            nn.Conv2d(128, 256, (1, 1), 2, 0),
            nn.BatchNorm2d(256)
        ))
        self.block_32 = resNetBlock(256, 1, 256)
        
        #4th set that shrinks data
        self.block_41 = resNetBlock(256, 2, 512, nn.Sequential(
            nn.Conv2d(256, 512, (1, 1), 2, 0),
            nn.BatchNorm2d(512)
        ))
        self.block_42 = resNetBlock(512, 1, 512)
        
        # actual layers to be called in forward function
        self.layers = nn.Sequential(
            nn.Conv2d(input_channels, 64, (7, 7), 2, 3),
            nn.BatchNorm2d(64),
            nn.ReLU(),
            nn.MaxPool2d(3, 2, 1),
            self.frontBlock,
            self.frontBlock,
            self.block_21,
            self.block_22,
            self.block_31,
            self.block_32,
            self.block_41,
            self.block_42,
            nn.AdaptiveAvgPool2d(1),
            nn.Flatten(),
            nn.Linear(512, out_num)
        )

    def forward(self, d):
        out = self.layers(d)
        return out
```

# <br><br><br><br><br><br><br><br><br> MODEL TESTING

### Function for testing

```python
#testing function (almost same as validation function)
def test(cnn_model, d_loader, l_function):
    cnn_model.eval()
    tst_loss = 0
    tst_correct = 0
    
    with torch.no_grad():
        for data, label in d_loader:
            data = data.to(device)
            label = label.to(device)
            
            estimate = cnn_model(data)
            tst_loss += l_function(estimate, label).item()
            tst_correct += (estimate.argmax(1) == label).type(torch.float).sum().item()
    
    # get averages / percentages, print out afterwards
    tst_avg_loss = tst_loss / len(d_loader)
    tst_correct /= len(d_loader.dataset)
    print("Testing Accuracy::", (100*tst_correct),end="")
    print("%\nAverage Testing Loss::",tst_avg_loss,"\n")
```
